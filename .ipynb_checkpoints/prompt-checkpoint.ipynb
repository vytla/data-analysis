### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

```python
#Analyze percentatge of Null values for each feature in Data Frame
print("Percentage of Missing Values in cloumns ")
count=0
for column in data.columns:
    percentage_miss = np.mean(data[column].isna())
    if percentage_miss > 0:
        count = count +1
        print(f'{column} :{round(percentage_miss * 100)}%')
if count == 0:
    print('No missing values found')
      
```

##### 2.2 Data perperation 
- 2.2.1 Identify and clean up duplicate rows

- 2.2.3 Rename features to meaningful names

```python
#2.2.1 Identify and cleanup duplicate rows 

total_rows = data.shape[0]
data=data.drop_duplicates()
print(f'Total {total_rows - data.shape[0]} duplicate rows cleaned up \n')

# 2.2.2 Rename last feature  'Y' to meaningful feature name to 'isCouponAccepted' and 
# have categorical values 'Yes' or 'No' for feature 'Y' 

data.rename({'Y' : 'isCouponAccepted'}, axis=1, inplace=True)
data['isCouponAccepted'] = data['isCouponAccepted'].map(lambda x: 'Yes' if x == 1 else 'No')

# 2.2.3 Rename features to meaningful name - time to timeOfTheDay 
data.rename({'time' : 'timeOfTheDay'}, axis=1, inplace=True)

# 2.2.4 Trim values for features of type string and print value counts
for col in data.columns:
    if isinstance(data[col],str):
        data[col] = data[col].str.strip()
    print('Column : ' + col)
    print(data[col].value_counts())
    print('\n')
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python

def getBar_Visits(barCount):
    if barCount in ("1~3"):
        return "Less than 3"
    elif barCount in ("gt8","4~8"):
        return "Greater than 3"
    else:
        return "Not Visted"

barcoupon_data['Bar_Visits'] = barcoupon_data['Bar'].map(getBar_Visits)


px.pie(barcoupon_data.groupby('Bar_Visits')['isCouponAccepted'].value_counts().reset_index().query('isCouponAccepted=="Yes" ' ).reset_index(),
       title='Coupon Acceptance Rate between Bar Visits less than 3 vs Greater than 3',
       values = 'count',
       names = 'Bar_Visits')
```

```python
bar_counts_df = barcoupon_data.groupby('Bar_Visits')['isCouponAccepted'].value_counts().reset_index().query('isCouponAccepted=="Yes" and Bar_Visits != "Not Visted"' ).reset_index()

px.histogram(bar_counts_df
            , x='Bar_Visits'
            , y='count'
            , color='Bar_Visits'
            , title ="Bar Coupons Used Less than 3 vs Greater than 3"
            ,nbins=10).update_layout(
            xaxis_title ="Bar Coupons Used" ,yaxis_title = "Total Count" )
```

```python

total_count = bar_counts_df.query('Bar_Visits == "Greater than 3"').iloc[0]['count'] +bar_counts_df.query('Bar_Visits == "Less than 3"').iloc[0]['count'] 
acceptance_greater_than_3 = round(bar_counts_df.query('Bar_Visits == "Greater than 3"').iloc[0]['count']/len(barcoupon_data) *100,2)
acceptance_less_than_3 = round(bar_counts_df.query('Bar_Visits == "Less than 3"').iloc[0]['count']/len(barcoupon_data) *100,2)

print(f'Acceptance Rate of Customers/Drivers who went to a bar 3 or more times is {acceptance_greater_than_3  }%')
print(f'Acceptance Rate of Customers/Drivers who went to a bar less than 3 times is {acceptance_less_than_3  }%')
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
bar_visit_more_than_Once = ['1~3','4~8','gt8']
ages_greater_than_25 = ['26','31','36','41','46','50plus']
def getBar_Age_25_Status(row):
   if row['Bar'] in bar_visit_more_than_Once and row['age'] in ages_greater_than_25 :
    return 'Yes'
   else:
    return 'No'
barcoupon_data['Bar_Age_25_Status'] = barcoupon_data.apply(getBar_Age_25_Status,axis=1)
```

```python
# Graph for Acceptance rate of drivers who go to bar more than once and age greater than 25
bar_25_pie = px.pie(barcoupon_data.groupby('Bar_Age_25_Status')['age'].count().reset_index()       
       , title='Acceptance rate for Drivers who go to Bar with age greater than 25'
       , values='age'
       , names='Bar_Age_25_Status').update_layout(legend={'title':'Bar visits more than once and older than 25'})
bar_25_pie.show()
```

```python

bar_by_25_hist= sns.histplot(barcoupon_data.query('age in @ages_greater_than_25').sort_values('age',ascending = True), x="age", hue="Bar_Age_25_Status", multiple="stack")
bar_by_25_hist.set(xlabel='Age')
plt.title('Acceptance of Bar coupons for Drivers Age is greater than 25\n')
plt.xticks(rotation=45)
graph_legend = bar_by_25_hist.get_legend()
graph_legend.set_bbox_to_anchor((1, 1))

graph_legend.set_title('Bar Coupons with Age > 25 ')
for t in graph_legend.get_texts():
    if t.get_text() == 'Yes':
        t.set_text('Coupon Accepted')
    else:
        t.set_text('Coupon Not Accepted')
plt.show()
```

###### Findings:
- From above pie chart,only 20.9% of Drivers/Customers who are older than 25 accepted Bar coupons. 
- Drivers/Customer with Age 26 and 31 years have higher acceptance rate for Bar coupons.

4. Compare the acceptance rate between those who used "Carry Out & take away" coupons 3 or fewer times a month to those who used more.

```python

def getCarryOut_Visits(carryOutCount):
    if carryOutCount in ("1~3"):
        return "Less than 3"
    elif carryOutCount in ("gt8","4~8"):
        return "Greater than 3"
    else:
        return "Not Used"

carryaway_coupon_data['carryOut_Visits'] = carryaway_coupon_data['CarryAway'].map(getCarryOut_Visits)


px.pie(carryaway_coupon_data.groupby('carryOut_Visits')['isCouponAccepted'].value_counts().reset_index().query('isCouponAccepted=="Yes" ' ).reset_index(),
       title='Coupon Acceptance Rate between Carryout coupon used less than 3 vs Greater than 3',
       values = 'count',
       names = 'carryOut_Visits')
```

```python
carryaway_counts_df = carryaway_coupon_data.groupby('carryOut_Visits')['isCouponAccepted'].value_counts().reset_index().query('isCouponAccepted=="Yes" and carryOut_Visits != "Not Used"' ).reset_index()

px.histogram(carryaway_counts_df
            , x='carryOut_Visits'
            , y='count'
            , color='carryOut_Visits'
            , title = 'Carry Out & take away coupons 3 or fewer times a month to those who used more'
            ,nbins=10).update_layout(
            xaxis_title ="Carry Out Coupons Used" ,yaxis_title = "Total Count" )
```

```python

total_count = carryaway_counts_df.query('carryOut_Visits == "Greater than 3"').iloc[0]['count'] +carryaway_counts_df.query('carryOut_Visits == "Less than 3"').iloc[0]['count'] 
acceptance_greater_than_3 = round(carryaway_counts_df.query('carryOut_Visits == "Greater than 3"').iloc[0]['count']/len(carryaway_coupon_data) *100,2)
acceptance_less_than_3 = round(carryaway_counts_df.query('carryOut_Visits == "Less than 3"').iloc[0]['count']/len(carryaway_coupon_data) *100,2)

print(f'Acceptance Rate of Customers/Drivers who used  Carry Out & take away coupons 3 or more times is {acceptance_greater_than_3  }%')
print(f'Acceptance Rate of Customers/Drivers who used  Carry Out & take away coupons less than 3 times is {acceptance_less_than_3  }%')
```
